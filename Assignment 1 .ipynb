```python
# 9. Write a Python program to convert user input from string to integer, float, and boolean types.

user_input = input("Enter something: ")

#convert to Integer
if user_input.isdigit():
    int_value = int(user_input)
    print("Integer value:", int_value)
else:
    print("Input is not a valid integer.")

#convert to Float
if user_input.count('.') == 1 and user_input.replace('.', '', 1).isdigit():
    float_value = float(user_input)
    print("Float value:", float_value)
elif user_input.isdigit():
    # In case it's an integer string, it can also be converted to float
    float_value = float(user_input)
    print("Float value:", float_value)
else:
    print("Input is not a valid float.")

# Convert to Boolean
if user_input:
    print("Boolean value: True")
else:
    print("Boolean value: False")
```

```python
# 13. Write a Python program to find the sum of all even numbers between 1 and 50.

sum = 0
for i in range(1, 51):
    if i % 2 == 0:
        sum += i
print("Total sum of even numbers from 1 to 50 is:", sum)
```

```python
# 14. Write a program to reverse a string using a while loop.

text = input("Enter a string: ")
reversed_text = ""
index = len(text) - 1
while index >= 0:
    reversed_text += text[index]
    index -= 1
print("Reversed string:", reversed_text)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a number: "))
factorial = 1
i = 1
while i <= num:
    factorial = factorial * i
    i = i + 12
print("Factorial is:", factorial)
```
